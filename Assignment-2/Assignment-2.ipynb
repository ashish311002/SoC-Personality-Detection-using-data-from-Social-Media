```python
data = pd.read_csv('task2.csv', header=0)
```

# Visualizing the Data

```python
#scatter plots
X = data.iloc[:, 0]

Y_temp = data.iloc[:, 1]
plt.scatter(X, Y_temp)
plt.show()

Y = np.log(data.iloc[:, 1])
plt.scatter(X, Y)
plt.show()
```

# Modelling

```python
m = 0
c = 0

L = 0.00001  # The learning Rate
iterations = 5000  # The number of iterations to perform gradient descent

n = 99 # Number of elements in X

# Performing Gradient Descent 
for i in range(iterations): 
    Y_pred = m*X + c  # The current predicted value of Y
    D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c
    
print (m, c)

#refernce: https://towardsdatascience.com/linear-regression-using-gradient-descent-97a6c8700931
```
